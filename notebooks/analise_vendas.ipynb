```python
import pandas as pd
import matplotlib.pyplot as plt

from scripts.limpeza_dados import carregar_dados, limpar_dados
```

```python
df = carregar_dados("../dados/vendas.csv")
df.head()
```

```python
df.info()
```

```python
df = limpar_dados(df)
df.head()
```

```python
df.describe()
```

```python
faturamento_produto = df.groupby("produto")["faturamento"].sum().sort_values(ascending=False)
faturamento_produto
```

```python
faturamento_produto.plot(kind="bar", title="Faturamento por Produto")
plt.xlabel("Produto")
plt.ylabel("Faturamento")
plt.show()
```

```python
df["data"] = pd.to_datetime(df["data"])
df["mes"] = df["data"].dt.to_period("M")

faturamento_mensal = df.groupby("mes")["faturamento"].sum()
faturamento_mensal
```

## Conclusão

A análise mostrou que os produtos da categoria Eletrônicos concentram o maior faturamento,
com destaque para notebooks e monitores. Também foi possível observar crescimento de vendas
nos meses analisados.
